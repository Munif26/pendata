# **Data Understanding**

# **Apa itu Data Understanding...?**
  Data Understanding adalah tahap awal dalam proses analisis data yang bertujuan untuk memahami karakteristik, struktur, dan kualitas data sebelum digunakan lebih lanjut. Pada tahap ini, data dikumpulkan dari berbagai sumber, kemudian dieksplorasi untuk mengetahui jumlah, tipe, serta distribusi nilai dalam setiap variabel. Evaluasi kualitas data juga dilakukan untuk mengidentifikasi masalah seperti data yang hilang, duplikasi, atau inkonsistensi yang dapat memengaruhi hasil analisis. Selain itu, dengan melakukan visualisasi dan analisis statistik dasar, kita dapat menemukan pola awal dan hubungan antarvariabel yang membantu dalam membentuk hipotesis atau asumsi yang dapat diuji lebih lanjut. Misalnya, dalam analisis harga rumah, memahami bagaimana faktor seperti lokasi, luas bangunan, dan jumlah kamar tidur memengaruhi harga dapat memberikan wawasan yang berguna sebelum membangun model prediktif. Dengan pemahaman data yang baik, proses analisis menjadi lebih akurat dan menghasilkan kesimpulan yang lebih terpercaya.

# **Macam-Macam Tipe Data!!!**
1. Numerical Data adalah data yang berupa angka dan dapat dihitung. Contoh: Usia, pendapatan.
2. Categorical Data adalah data yang menggambarkan kategori atau kelompok. Contoh: jenis kelamin, kategori produk.
3. Ordinal Data adalah data yang memiliki urutan atau tingkatan. Contoh: tingkat pendidikan

Berikut lebih jelasnya :
Tipe-tipe data dalam data understanding dapat diklasifikasikan berdasarkan beberapa aspek, yaitu sifatnya, skala pengukuran, bentuknya, dan sumbernya.  

Berdasarkan sifatnya, data dapat dibagi menjadi dua jenis utama. Data numerik merupakan data yang berbentuk angka dan dapat dihitung atau diukur, seperti usia atau pendapatan seseorang. Sementara itu, data kategorikal atau kualitatif adalah data yang berbentuk kategori atau label, misalnya jenis kelamin atau warna mata.  

Berdasarkan skala pengukurannya, data dapat dikategorikan menjadi empat jenis. Data nominal merupakan data dalam bentuk kategori tanpa adanya urutan tertentu, seperti warna atau jenis kelamin. Data ordinal memiliki kategori yang memiliki urutan, tetapi perbedaan antar kategorinya tidak memiliki jarak yang pasti, contohnya tingkat pendidikan atau tingkat kepuasan pelanggan. Data interval adalah data numerik dengan jarak tetap antara nilai, tetapi tidak memiliki nol absolut, seperti suhu dalam satuan Celsius atau Fahrenheit. Terakhir, data rasio adalah data numerik yang memiliki nol absolut, sehingga memungkinkan perbandingan rasio, seperti berat badan atau pendapatan seseorang.  

Berdasarkan bentuknya, data dapat diklasifikasikan menjadi data diskrit dan data kontinu. Data diskrit adalah data yang hanya memiliki nilai tertentu dan tidak dapat dibagi menjadi pecahan, seperti jumlah anak dalam sebuah keluarga atau jumlah mobil yang dimiliki seseorang. Di sisi lain, data kontinu adalah data yang dapat memiliki nilai dalam rentang tertentu, termasuk nilai pecahan, seperti tinggi badan atau waktu perjalanan.  

Berdasarkan sumbernya, data dibagi menjadi data primer dan data sekunder. Data primer adalah data yang dikumpulkan langsung dari sumbernya, misalnya melalui survei, wawancara, atau eksperimen. Sebaliknya, data sekunder adalah data yang diperoleh dari sumber yang sudah ada sebelumnya, seperti laporan penelitian, database perusahaan, atau publikasi resmi.  

## **Bagaimana Data Understanding dalam Data Mining...?**
  Data Understanding dalam Data Mining merupakan tahap awal yang bertujuan untuk memahami karakteristik data sebelum dilakukan analisis lebih lanjut. Langkah ini mencakup pengumpulan data dari berbagai sumber, deskripsi data untuk mengetahui tipe dan jumlah data yang tersedia, serta eksplorasi data melalui teknik statistik dan visualisasi guna mengidentifikasi pola, distribusi, dan potensi anomali seperti outlier. Selain itu, tahap ini juga melibatkan verifikasi kualitas data, termasuk menangani nilai yang hilang, duplikasi, serta inkonsistensi yang dapat memengaruhi hasil analisis. Beberapa teknik yang umum digunakan dalam proses ini meliputi statistik deskriptif, analisis korelasi, dan berbagai metode visualisasi seperti histogram, scatter plot, serta heatmap. Pemahaman yang baik terhadap data pada tahap ini sangat penting karena akan memengaruhi kualitas model yang dibangun di tahap selanjutnya, seperti preprocessing dan modeling. Dengan demikian, Data Understanding menjadi fondasi utama dalam proses Knowledge Discovery in Databases (KDD) untuk memastikan analisis yang akurat dan efektif.

## **2. Sampling Data**
  Sampling data dalam Data Understanding adalah teknik yang digunakan untuk mengambil sebagian data dari keseluruhan dataset guna mewakili karakteristik populasi secara akurat. Sampling dilakukan karena dalam banyak kasus, jumlah data yang sangat besar membuat analisis langsung terhadap seluruh dataset menjadi tidak efisien dan memakan banyak sumber daya. Dengan melakukan sampling, analisis awal dapat dilakukan lebih cepat tanpa kehilangan informasi penting dari data. Terdapat beberapa metode sampling yang umum digunakan, di antaranya random sampling, di mana setiap data memiliki peluang yang sama untuk dipilih; stratified sampling, yang membagi data ke dalam kelompok-kelompok (strata) dan mengambil sampel dari setiap kelompok secara proporsional; serta systematic sampling, yang memilih data dengan pola tertentu, misalnya setiap data ke-10 dari dataset. Selain itu, ada juga cluster sampling, yang membagi data menjadi beberapa kelompok dan mengambil sampel dari kelompok tertentu secara acak.  

Sampling dalam Data Understanding bertujuan untuk menghemat waktu dan sumber daya, namun tetap mempertahankan representasi yang baik dari keseluruhan dataset. Dengan menggunakan teknik sampling yang tepat, proses eksplorasi data dapat dilakukan dengan lebih efisien, memungkinkan deteksi awal pola, anomali, serta potensi masalah dalam data sebelum masuk ke tahap pemrosesan lebih lanjut. ## **3. Profiling Data**
Profiling data dalam Data Understanding adalah proses menganalisis dan mengidentifikasi karakteristik dari data untuk memahami struktur, kualitas, serta pola yang terkandung di dalamnya. Tujuan utama dari data profiling adalah untuk mendeteksi ketidakkonsistenan, nilai yang hilang, duplikasi, serta memahami distribusi data yang ada. Dengan melakukan profiling, peneliti atau analis dapat mengevaluasi apakah data yang tersedia memenuhi syarat untuk digunakan dalam analisis lebih lanjut atau memerlukan pembersihan dan transformasi tambahan. Proses data profiling melibatkan beberapa langkah utama. Pertama, analisis struktur data, yaitu pemeriksaan atribut, tipe data, serta format yang digunakan. Kedua, analisis konten data, yang mencakup pengecekan statistik deskriptif seperti mean, median, modus, distribusi nilai, serta outlier yang ada dalam dataset. Ketiga, analisis hubungan antar-data, di mana dilakukan evaluasi korelasi antar variabel, deteksi duplikasi, serta identifikasi anomali atau inkonsistensi dalam data.

Dengan melakukan data profiling, organisasi atau analis data dapat lebih mudah mengidentifikasi potensi permasalahan yang dapat mempengaruhi kualitas hasil analisis. Selain itu, tahap ini juga membantu dalam proses pengambilan keputusan terkait strategi preprocessing data yang harus dilakukan sebelum masuk ke tahap pemodelan dalam Data Mining. Profiling data menjadi langkah penting dalam memastikan bahwa data yang digunakan bersih, valid, dan siap untuk menghasilkan insight yang akurat.

## **5. Korelasi dan Asosiasi**
Dalam Data Understanding, korelasi dan asosiasi merupakan dua konsep penting yang digunakan untuk menganalisis hubungan antar variabel dalam suatu dataset. Korelasi mengukur sejauh mana dua variabel numerik memiliki hubungan linear, yang biasanya dinyatakan dengan koefisien korelasi Pearson (r). Jika nilai korelasi positif, berarti kedua variabel bergerak searah, sedangkan korelasi negatif menunjukkan hubungan berlawanan, dan nilai nol menandakan tidak adanya hubungan linear. Selain Pearson, terdapat juga korelasi Spearman dan Kendall yang digunakan untuk hubungan non-linear. Di sisi lain, asosiasi lebih berfokus pada hubungan antara item dalam dataset, terutama dalam market basket analysis atau analisis pola transaksi. Teknik Association Rule Mining, seperti algoritma Apriori dan FP-Growth, digunakan untuk menemukan pola dengan mengukur metrik seperti support, confidence, dan lift, yang menentukan seberapa kuat hubungan antara item dalam data. Misalnya, jika pelanggan membeli roti, ada kemungkinan besar mereka juga membeli selai. Meskipun keduanya bertujuan untuk menemukan hubungan dalam data, korelasi lebih relevan untuk variabel numerik dengan hubungan linear, sedangkan asosiasi lebih sering digunakan untuk menemukan pola dalam data kategorikal atau transaksi. Memahami kedua konsep ini sangat penting dalam eksplorasi data agar analisis yang dilakukan lebih akurat dan bermanfaat untuk pengambilan keputusan berbasis data.


## **6. Eksplorasi Data**

Eksplorasi data merupakan proses awal dalam analisis data untuk memahami karakteristik, distribusi, dan pola dalam dataset yang ada dan akan menghasilkan deskripsi statistik seperti mean, median, modus, standar deviasi,minimum, dan maximum dengan menggunakan visualisasi data bisa dalam bentuk grafik histogram scatter plot untuk melihat pola atau trend awalnya.


## **7. Visualisasi Data**

**7.1 Bar Chart**

```python
#Kode ini digunakan untuk menganalisis dan memvisualisasikan distribusi panjang sepal (sepal length) dalam dataset Iris berdasarkan class
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Memuat dataset Iris dari Seaborn
combined_df = sns.load_dataset("iris")

#Pastikan nama kolom sesuai dengan dataset
sepal_length_col = "sepal_length"
class_col = "species"

#Pilih warna khusus untuk setiap class
custom_palette = {
    "setosa": "red",
    "versicolor": "green",
    "virginica": "blue"  #Menggunakan biru agar lebih terlihat daripada putih
}

#Langkah selanjutnya menghitung statistik dasar
mean_sepal_length = combined_df[sepal_length_col].mean()
median_sepal_length = combined_df[sepal_length_col].median()
min_sepal_length = combined_df[sepal_length_col].min()
max_sepal_length = combined_df[sepal_length_col].max()
mode_sepal_length = combined_df[sepal_length_col].mode()[0]

#Membuat plot distribusi dengan warna yang sudah ditentukan
plt.figure(figsize=(10, 4))
sns.histplot(data=combined_df, x=sepal_length_col, hue=class_col, bins=20, kde=True, alpha=0.6, palette=custom_palette)

#Menambahkan garis rata-rata, median, min, dan max
plt.axvline(mean_sepal_length, color="black", linestyle="--", label=f"Mean: {mean_sepal_length:.2f}")
plt.axvline(mode_sepal_length, color="blue", linestyle="--", label=f"Mode: {mode_sepal_length:.2f}")
plt.axvline(median_sepal_length, color="yellow", linestyle="--", label=f"Median: {median_sepal_length:.2f}")
plt.axvline(min_sepal_length, color="brown", linestyle="--", label=f"Min: {min_sepal_length:.1f}")
plt.axvline(max_sepal_length, color="purple", linestyle="--", label=f"Max: {max_sepal_length:.1f}")

#Tambahkan label dan judul
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Distribusi Sepal Length Berdasarkan Class (Red, Green, Blue)")
plt.legend(title="Statistics")

#Untuk menampilkan plot
plt.show()
```

# **8. Identifikasi Masalah Data**
Identifikasi Masalah dalam Eksplorasi Data: Deteksi Outlier pada Data Iris Sebelum melakukan eksplorasi data, langkah pertama yang harus dilakukan adalah Identifikasi Masalah. Proses ini bertujuan untuk memahami permasalahan dalam dataset dan menentukan langkah-langkah yang diperlukan agar analisis lebih akurat dan efisien. Salah satu aspek penting dalam identifikasi masalah adalah mendeteksi outlier, yaitu data yang menyimpang secara signifikan dari sebagian besar data lainnya. Outlier dapat disebabkan oleh kesalahan pengukuran, pencatatan, atau memang merupakan fenomena khusus yang perlu dianalisis lebih lanjut. Dalam konteks dataset Iris, kita dapat menggunakan teknik visualisasi seperti boxplot untuk mengidentifikasi outlier dalam variabel numerik, seperti sepal length, sepal width, petal length, dan petal width. Outlier ditandai sebagai titik-titik yang berada jauh dari rentang normal data. Dengan mendeteksi dan menangani outlier, kita dapat memastikan bahwa data yang digunakan dalam analisis lebih bersih, sehingga hasil eksplorasi data dan model prediksi yang dibuat nantinya menjadi lebih akurat dan efisien.

```python
# Pastikan nama kolom sesuai dengan dataset
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset Iris
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Tambahkan kolom class jika diperlukan
df["class"] = ["Iris-setosa"] * 50 + ["Iris-versicolor"] * 50 + ["Iris-virginica"] * 50

# Simpan ke dalam filtered_df2
filtered_df2 = df

df = filtered_df2

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # Kuartil pertama (Q1)
    Q3 = df[column].quantile(0.75)  # Kuartil ketiga (Q3)
    IQR = Q3 - Q1  # Interquartile Range

    # Batas bawah dan atas untuk mendeteksi outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Menyaring data yang berada di luar batas sebagai outlier
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    return outliers[[column]]

# Gunakan nama kolom yang benar
ptl_outliers = detect_outliers(df, "petal length (cm)")
ptw_outliers = detect_outliers(df, "petal width (cm)")
spl_outliers = detect_outliers(df, "sepal length (cm)")
spw_outliers = detect_outliers(df, "sepal width (cm)")

# Menampilkan hasil
print(f"Outliers pada kolom Petal Length:")
print(ptl_outliers)
print()
print(f"Outliers pada kolom Petal Width:")
print(ptw_outliers)
print()
print(f"Outliers pada kolom Sepal Length:")
print(spl_outliers)
print()
print(f"Outliers pada kolom Sepal Width:")
print(spw_outliers)
print()
```

# **9. Validasi Data**

Validasi data merupakan langkah krusial dalam tahap pemahaman data (data understanding) yang bertujuan untuk memastikan bahwa data yang digunakan relevan, akurat, dan mencerminkan realitas yang ingin dipelajari. Proses ini dilakukan dengan berbagai cara, seperti memeriksa kelengkapan data, memastikan format dan tipe data sesuai, serta menguji konsistensi antar-kolom.
