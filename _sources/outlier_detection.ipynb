#**Outliers Detection Menggunakan K-NN**

#**Apa itu Outliers Detections...?**

Outlier detection adalah proses mengidentifikasi data yang menyimpang secara signifikan dari pola umum dalam suatu dataset. Outlier merupakan data yang memiliki nilai jauh berbeda dibandingkan dengan sebagian besar data lainnya, yang dapat disebabkan oleh kesalahan pengukuran, variasi alami, atau kejadian langka. Deteksi outlier sangat penting dalam berbagai bidang, seperti analisis keuangan untuk mendeteksi transaksi penipuan, bidang kesehatan untuk mengidentifikasi hasil medis yang abnormal, serta pemeliharaan prediktif untuk mencegah kerusakan mesin. Ada beberapa metode yang umum digunakan untuk mendeteksi outlier, di antaranya metode statistik seperti Z-Score dan Interquartile Range (IQR) yang menghitung distribusi data untuk menentukan apakah suatu nilai termasuk outlier. Selain itu, metode berbasis machine learning seperti K-Means Clustering, Isolation Forest, dan DBSCAN dapat digunakan untuk mengenali pola outlier secara lebih kompleks. Visualisasi data seperti boxplot dan scatter plot juga sering dimanfaatkan untuk mendeteksi outlier secara intuitif. Dengan mendeteksi outlier secara efektif, kualitas analisis data dapat ditingkatkan, sehingga menghasilkan wawasan yang lebih akurat dan dapat diandalkan.


#**Cara mendeteksi Outliers**

Ada banyak cara untuk mendeteksi Outliers salah satunya adalah dengan menggunakan yaitu Metode K-Nearest Neighbors (KNN) dapat digunakan untuk mendeteksi outliers dengan cara menganalisis kedekatan setiap titik data dengan tetangganya. Dalam konteks deteksi outlier, KNN menghitung jarak antara satu titik data dengan k tetangga terdekatnya. Jika suatu data memiliki jarak rata-rata yang jauh lebih besar dibandingkan dengan titik data lainnya, maka data tersebut dapat dianggap sebagai outlier. Proses ini bekerja berdasarkan asumsi bahwa data normal cenderung berdekatan satu sama lain dalam suatu ruang fitur, sedangkan outlier akan berada jauh dari kelompok data utama. Untuk mendeteksi outlier menggunakan KNN, langkah pertama adalah menentukan jumlah tetangga terdekat (k), lalu menghitung rata-rata jarak titik ke tetangganya. Setelah itu, ditentukan threshold atau ambang batas tertentu, di mana data yang memiliki nilai rata-rata jarak lebih tinggi dari threshold akan diklasifikasikan sebagai outlier. Metode KNN ini efektif digunakan dalam dataset dengan distribusi yang tidak diketahui dan dapat menangani data berdimensi tinggi, tetapi memiliki kelemahan dalam hal komputasi karena memerlukan perhitungan jarak untuk setiap pasangan data, terutama jika jumlah data sangat besar.

#**Apa itu Metode K-Nearest Neighbors (KNN)...?**
K-Nearest Neighbors (KNN) adalah algoritma dalam machine learning yang digunakan untuk klasifikasi dan regresi dengan menentukan kedekatan suatu data terhadap tetangga terdekatnya. Algoritma ini termasuk dalam metode lazy learning, karena tidak memerlukan proses pelatihan model secara eksplisit, melainkan menyimpan seluruh data pelatihan dan melakukan perhitungan setiap kali ada data baru yang perlu diklasifikasikan atau diprediksi. Cara kerja KNN dimulai dengan menentukan jumlah tetangga terdekat (k), kemudian menghitung jarak antara data baru dan setiap data dalam dataset menggunakan metrik seperti Euclidean Distance, Manhattan Distance, atau Minkowski Distance. Setelah itu, dipilih k data dengan jarak terdekat, dan prediksi dilakukan berdasarkan mayoritas kelas (untuk klasifikasi) atau rata-rata nilai (untuk regresi). Kelebihan utama KNN adalah kesederhanaannya, fleksibilitasnya dalam berbagai jenis data, dan kemampuannya digunakan baik untuk klasifikasi maupun regresi. Namun, KNN memiliki kelemahan dalam hal efisiensi, terutama pada dataset besar, karena harus menghitung jarak ke seluruh data. Selain itu, pemilihan nilai k yang tepat sangat penting untuk mendapatkan hasil yang optimal. Algoritma ini banyak digunakan dalam aplikasi seperti pengenalan pola, sistem rekomendasi, serta deteksi anomali dalam data transaksi keuangan.

#**Cara kerja K-Nearest Neighbors (KNN)**
### **Cara Kerja K-Nearest Neighbors (KNN) Secara Detail**  

K-Nearest Neighbors (KNN) bekerja dengan prinsip dasar bahwa data yang memiliki karakteristik serupa cenderung berada dalam jarak yang berdekatan. Algoritma ini tidak memiliki tahap pelatihan eksplisit seperti model machine learning lainnya, tetapi menyimpan seluruh data pelatihan dan melakukan prediksi dengan mencari k tetangga terdekat dari data baru yang diberikan. Berikut adalah langkah-langkah kerja KNN secara rinci:  

---

### **1. Menentukan Nilai k (Jumlah Tetangga Terdekat)**  
- Nilai **k** merupakan jumlah tetangga yang akan dipertimbangkan dalam proses klasifikasi atau regresi.  
- Pemilihan **k** yang terlalu kecil bisa menyebabkan model terlalu sensitif terhadap noise, sedangkan **k** yang terlalu besar bisa menyebabkan model terlalu umum dan kurang akurat.  
- Biasanya, **k dipilih sebagai bilangan ganjil** untuk menghindari kemungkinan hasil seri dalam klasifikasi.  

---

### **2. Menghitung Jarak Antara Data Baru dengan Semua Data dalam Dataset**  
Setelah k ditentukan, langkah berikutnya adalah menghitung jarak antara data yang ingin diklasifikasikan (data uji) dengan semua data dalam dataset (data pelatihan). Ada beberapa metode pengukuran jarak yang sering digunakan:

---

### **3. Menentukan k Data dengan Jarak Terdekat**  
- Setelah jarak antara data uji dan seluruh data dalam dataset dihitung, langkah selanjutnya adalah memilih **k data dengan jarak terkecil**.  
- Data ini disebut sebagai **tetangga terdekat (nearest neighbors)** dari data uji.  

---

### **4. Menentukan Prediksi Berdasarkan k Tetangga Terdekat**  
Terdapat dua kemungkinan jenis prediksi yang dilakukan oleh KNN:  

-> **Untuk Klasifikasi:**  
- KNN melihat **mayoritas kelas** dari k tetangga terdekat.  
- Jika mayoritas dari k tetangga berada dalam satu kelas tertentu, maka data uji diklasifikasikan ke dalam kelas tersebut.  
- Misalnya, jika dari 5 tetangga terdekat, 3 termasuk dalam kelas "A" dan 2 dalam kelas "B", maka data uji diklasifikasikan sebagai kelas "A".  

-> **Untuk Regresi:**  
- KNN menghitung **rata-rata nilai target** dari k tetangga terdekat.  
- Misalnya, jika k = 3 dan nilai target dari tiga tetangga adalah 50, 55, dan 60, maka nilai prediksi akan dihitung sebagai:  
   50 + 55 + 60 / 3 = 55

---

### **5. Menampilkan Hasil Prediksi**  
Setelah prediksi ditentukan, algoritma KNN mengembalikan hasilnya, baik dalam bentuk kelas (jika klasifikasi) maupun dalam bentuk nilai numerik (jika regresi).  

---

### **Contoh Ilustrasi Cara Kerja KNN dalam Klasifikasi**  
Misalkan kita memiliki dataset dengan dua kelas: **Biru** dan **Merah**, serta sebuah titik data baru yang ingin diklasifikasikan.  

1. **Menentukan nilai k** → Misalnya, kita pilih **k = 3**.  
2. **Menghitung jarak** antara titik data baru dan semua titik lainnya dalam dataset.  
3. **Memilih 3 titik terdekat** → Misalkan 2 titik adalah **Biru** dan 1 titik adalah **Merah**.  
4. **Menentukan kelas mayoritas** → Karena mayoritas dari 3 tetangga adalah **Biru**, maka titik data baru diklasifikasikan sebagai **Biru**.  

#**Detection Outlier menggunakan Metode K-Nearest Neighbors (KNN)**

### **Deteksi Outliers dengan Metode K-Nearest Neighbors (KNN)**  

Metode **K-Nearest Neighbors (KNN)** dapat digunakan untuk mendeteksi **outliers**, yaitu data yang memiliki perbedaan signifikan dibandingkan dengan sebagian besar data lainnya dalam suatu dataset. Dalam pendekatan ini, kita menentukan apakah suatu titik data adalah outlier dengan melihat **jarak rata-rata** antara titik tersebut dan k tetangga terdekatnya. Jika jarak rata-rata jauh lebih besar dibandingkan dengan data lainnya, maka titik tersebut dapat dianggap sebagai **outlier**.  

---

### **Langkah-Langkah Deteksi Outliers dengan KNN**  

1. **Menentukan Nilai k (Jumlah Tetangga Terdekat)**  
   - Nilai **k** menentukan berapa banyak tetangga yang akan dipertimbangkan saat menganalisis suatu titik.  
   - Jika **k terlalu kecil**, metode ini menjadi lebih sensitif terhadap noise. Jika **k terlalu besar**, metode ini bisa gagal mendeteksi outliers dengan baik.  

2. **Menghitung Jarak Setiap Titik ke Tetangga Terdekatnya**  
   - Untuk setiap titik dalam dataset, dihitung jaraknya ke semua titik lainnya.  
   - Jarak dapat dihitung menggunakan beberapa metode, seperti **Euclidean Distance** atau **Manhattan Distance**.  

3. **Menghitung Rata-rata Jarak ke k Tetangga Terdekat**  
   - Setelah mendapatkan jarak ke semua titik, diambil **k tetangga terdekat** dan dihitung rata-rata jaraknya.  
   - Jika suatu titik memiliki rata-rata jarak yang jauh lebih besar dibandingkan dengan kebanyakan titik lainnya, kemungkinan besar titik tersebut adalah **outlier**.  

4. **Menentukan Ambang Batas (Threshold) untuk Outlier**  
   - Untuk membedakan titik normal dan outlier, digunakan metode statistik seperti:  
     - **Persentil (Percentile Method)** → Misalnya, titik dengan nilai rata-rata jarak dalam **10% tertinggi** dianggap sebagai outlier.  
     - **Z-score Method** → Jika nilai rata-rata jarak suatu titik sangat jauh dari rata-rata dataset, maka titik tersebut diklasifikasikan sebagai outlier.  

5. **Menandai dan Menganalisis Outliers**  
   - Data yang melebihi threshold dianggap sebagai outlier dan perlu dianalisis lebih lanjut.  
   - Dalam beberapa kasus, outliers dapat dihapus atau diolah ulang tergantung pada kebutuhan analisis data.  

---

### **Contoh Ilustrasi**  
Misalkan kita memiliki dataset dengan banyak titik data, dan kita ingin mencari outliers menggunakan KNN:  
1. Kita menentukan **k = 3**, artinya kita akan mempertimbangkan **3 tetangga terdekat** dari setiap titik.  
2. Untuk setiap titik, kita menghitung jaraknya ke semua titik lainnya dan memilih **3 tetangga terdekat**.  
3. Jika rata-rata jarak suatu titik ke **3 tetangga terdekatnya** sangat besar dibandingkan dengan titik lainnya, maka titik tersebut adalah **outlier**.  

Contohnya, jika sebagian besar data memiliki jarak rata-rata ke tetangganya sekitar **5-10 unit**, tetapi ada satu titik dengan rata-rata jarak **50 unit**, maka titik tersebut kemungkinan besar adalah **outlier**.
